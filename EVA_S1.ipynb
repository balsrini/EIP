```
# import the necessary packages
import numpy as np
from urllib.request import urlopen
import cv2
from matplotlib import pyplot as plt

# METHOD #1: OpenCV, NumPy, and urllib
def url_to_image(url):
	# download the image, convert it to a NumPy array, and then read
	# it into OpenCV format
	resp = urlopen(url)
	image = np.asarray(bytearray(resp.read()), dtype="uint8")
	image = cv2.imdecode(image, cv2.IMREAD_COLOR)
 
	# return the image
	return image
```

```
image = url_to_image("https://fontsarena-cd5e.kxcdn.com/wp-content/uploads/2019/04/helvetica-now-font-400x364.png")
```

```
from google.colab.patches import cv2_imshow
cv2_imshow(image)
```

```
edges = cv2.Canny(image,100,200)
```

```
cv2_imshow(edges)
```

```
#kernel = np.ones((3,3),np.float32)/25
kernel = np.float32([[-1,0,1],[-1,0,1],[-1,0,1]])

dst = cv2.filter2D(image,-1,kernel)
cv2_imshow(dst)
```

```
#Creating a horizontal kernel. 
horizontalKernel = np.float32([[-1,-2,-1],[0,0,0],[1,2,1]])
hDst = cv2.filter2D(image,-1,horizontalKernel)
cv2_imshow(hDst)

#creating a 45 degree edge
angleDetector = np.float32([[0,1,2],[-1,0,1],[-2,-1,0]])
angleDst = cv2.filter2D(image,-1,angleDetector)
cv2_imshow(angleDst)

#Blur kernel
positiveLaplace = np.float32([[0,1,0],[1,-4,1],[0,1,0]])
blurImg = cv2.filter2D(image,-1,positiveLaplace)
cv2_imshow(blurImg)

#sharpen kernel
sharpenKernel = np.float32([[0,-1,0],[-1,5,-1],[0,-1,0]])
sharpenImg = cv2.filter2D(image,-1,sharpenKernel)
cv2_imshow(sharpenImg)

#identity Function
idKernel = np.float32([[0,0,0],[0,1,0],[0,0,0]])
idImg = cv2.filter2D(image,-1,idKernel)
cv2_imshow(idImg)
```
